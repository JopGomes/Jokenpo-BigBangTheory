```python

options=['pedra','papel','tesoura','lagarto','spock']

winners={ #pedra ganha de tesoura
    "pedra":["tesoura","lagarto"],
    "lagarto":["papel","spock"],
    "spock":["tesoura","pedra"],
    "tesoura":["lagarto","papel"],
    "papel":["spock","pedra"]
}

losers={ #pedra perde para spock
    "pedra":["spock","papel"],
    "lagarto":["pedra","tesoura"],
    "spock":["papel","lagarto"],
    "tesoura":["spock","pedra"],
    "papel":["tesoura","lagarto"]
}

tradutor={
    "0":"pedra",
    "1":"spock",
    "2":"papel",
    "3":"lagarto",
    "4":"tesoura",
    "pedra":"0",
    "spock":"1",
    "papel":"2",
    "lagarto":"3",
    "tesoura":"4"
}

def result(mine,opponent):
    if(opponent in winners[mine]):
        return 'Vitoria'
    elif(mine==opponent):
        return 'Empate'
    else:
        return 'Derrota'
def loser(mine,opponent):
    if(opponent in losers[mine]):
        return True
    else:
        return False
```

```python
import statistics
from reportlab.pdfgen import canvas
import matplotlib.pyplot as plt
from pymongo import MongoClient


msg = ['pedra','papel','pedra','spock','lagarto','tesoura']
def nextPlay(msgOpponent):
    moda = statistics.mode(msgOpponent)
    if(moda):
        msgAu = [i for i in msgOpponent if i != moda]
        moda2 = statistics.mode(msgAu)
        ganhaDaModa = losers[moda]
        if(moda2):
            for i in ganhaDaModa:
                if( not bool(loser(i,moda2))):
                    return i
        else:
            return ganhaDaModa[0]
    else:
        return "spock"
def printAnswer(msgOpp,msgAnswer):
    k=0
    log=''
    for i in msgAnswer:
        j=msgOpp[k]
        k+=1
        log+=f'Na rodada {k}:\n O adversario jogou {j} e foi jogado pelo bot python {i} resultando em {result(i,j)}\n'
    log+='\n\n'
    return log

def GeneratePDF(msgOpp,msgAnswer):
    nome_pdf = 'Resultados'
    try:
        resultados=[]
        img=[]
        pdf = canvas.Canvas('{}.pdf'.format(nome_pdf))
        x = 700
        k=0
        for i in msgOpp:
            j=msgOpp[k]
            k+=1
            x -= 20
            pdf.drawString(247,x, 'Rodada nº {}- {}   x   {}   :  {}'.format(k, i,j,result(i,j) ) )
            resultados.append(result(i,j))
            
        pdf.setTitle(nome_pdf)
        pdf.setFont("Helvetica-Oblique", 14)
        pdf.drawString(245,750, 'Lista de Resultados')
        pdf.drawString(80,750, 'Lista de Gráficos')
        pdf.setFont("Helvetica-Bold", 10)
        pdf.drawString(245,724, 'Rodada - Jogada oponente  x  Jogada bot python : Resultado')
        
        
        plt.title('Frequência de valores para o oponente')
        plt.xlabel('Classe')
        plt.ylabel('Numero de vezes')
        plt.hist(msgOpp)
        plt.savefig('opp.png')
        img.append('opp.png')
        plt.show()
        
        plt.title('Frequência de valores para o bot')
        plt.xlabel('Classe')
        plt.ylabel('Numero de vezes')
        plt.hist(msgOpp)
        plt.savefig('bot.png')
        img.append('bot.png')
        plt.show()
        
        plt.title('Resultado geral')
        plt.xlabel('Classe')
        plt.ylabel('Numero de vezes')
        plt.hist(resultados)
        plt.savefig('resultado.png')
        img.append('resultado.png')
        plt.show()
        
        x_start=0
        y_start=x-100
        for img_file in img:
            pdf.drawImage(img_file, x_start, y_start, width=240, preserveAspectRatio=True, mask='auto')
            y_start-=200
        pdf.save()
        
        print('{}.pdf criado com sucesso!'.format(nome_pdf))
    except:
        print('Erro ao gerar {}.pdf'.format(nome_pdf))

def get_database():
 
   # Provide the mongodb atlas url to connect python to mongodb using pymongo
    CONNECTION_STRING = "mongodb+srv://user:pass@cluster.mongodb.net/NomeDataBase"
 
   # Create a connection using MongoClient. You can import MongoClient or use pymongo.MongoClient
    client = MongoClient(CONNECTION_STRING)
 
   # Create the database for our example (we will use the same database throughout the tutorial
    return client['user_shopping_list']

        
def saveMongo(msgOpp,msgAnswer):
    k=0
    a={}
    #dbname = get_database()
    #rodada = dbname["Rodada"]
    for i in msgOpp:
        j=msgOpp[k]
        k+=1
        key = f'item_{k}'
        value = {
        "Rodada":k,
        "Oponente" : j,
        "BotPython" : i,
        "Resultado": result(i,j),
        }
        a[key]=value
        #rodada.insert_one(a[key])
    print(a)
```

```python
from socket import *
HOST,PORT = "172.15.1.211",40000

msgOpponent = []
msgAnswer = []
continua = True
with socket(AF_INET,SOCK_STREAM) as s:
    s.bind((HOST,PORT))
    s.listen()
    conn,addr=s.accept()
    with conn:
        print(f"Endereço do Cliente {addr}")
        while continua:
            cliente = conn.recv(1024)
            data= nextPlay(msgOpponent)
            msgOpponent.append(tradutor[cliente.decode()])
            msgAnswer.append(data)
            conn.sendall(tradutor[data])
            printAnswer(msgOpponent,msgAnswer)
            if(len(msgOpponent)==15)break
        GeneratePDF(msgOpponent,msgAnswer)
            
HOST,PORT = "172.15.3.50",40000

with socket(AF_INET,SOCK_STREAM) as s:  
    s.connect((HOST,PORT))
    while continua:
        answer = nextPlay(msgOpponent)
        s.sendall(tradutor[answer].encode())
        opp = s.recv(1024)
        msgAnswer.append(answer)
        msgOpponent.append(tradutor[opp])
        printAnswer(msgOpponent,msgAnswer)
        if(len(msgOpponent)==15)break
    GeneratePDF(msgOpponent,msgAnswer)
```
